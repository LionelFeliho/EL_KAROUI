## Exercice 1: Pricing d'options européennes dans le modèle binomial

On s'intéresse au calcul du prix d'un payoff Européen $\Psi(S_{t_n})$ dans le modèle binomial avec paramètres $ u> d$, valeur initial de l'actif $S_0$, et en présence d'un taux d'intéret $r$ sur chaque période: $1$ Euro en $t_i$ devient $1+r$ Euros en $t_{i+1}$.

#### $\blacktriangleright$ Portefeuille de couverture

Comme démontré dans le cours, en supposant un modèle binomial, pour chaque payoff $\Psi(S_{t_n})$ il existe un unique portefeuille autofinançant $(V_{t_i})_{i=0,\dots,n}$ qui réplique exactement le payoff à maturité. 

Ce portefeuille s'écrit à chaque date $t_i$ comme une fonction $v(t_i, S_{t_i})$ de la valeur courante du sous-jacent en $t_i$.

### Questions 1 et 2: nous répondons ci-dessous aux deux questions simultanément, en utilisant la meme fonction pour calculer le prix $V_{t_i} = v(t_i, S_{t_i})$ et le delta $\delta(t_i, S_{t_i})$ de l'option à chaque date $t_i$ .

Tout d'abord, nous définissons une fonction Python qui renvoie la fonction valeur du portefeuille $v(t_i, \cdot)$ à la date $t_i$ et son delta $\delta(t_i, \cdot)$ à partir de la connaissance de la fonction valeur $v(t_{i+1}, \cdot)$ à la date $t_{i+1}$ suivante.

```python
import numpy as np
```

```python
def recursion(r, u, d, fonction_prix_t_i_plus_one, delta_flag=0):
    """
    On effectue une itération de la date t_{i+1} à la date t_i.
    
    Paramètres
    + r, u, d: les paramètres du modèle binomial 
    + fonction_prix_t_i_plus_one (une fonction Python):
        La fonction de prix v(t_{i+1}, .) à la date t_{i+1}.
        Doit prendre comme argument la valeur S_{t_{i+1}} du sous-jacent à la date t_{i+1}.
    
    Output:
    + Une fonction Python, la fonction de prix v(t_i, .) à la date t_i.
      Doit prendre comme argument la valeur S_{t_i} du sous-jacent.
            
    + Si delta_flag: renvoie une autre fonction Python, le delta delta(t_i, .) du portefeuille à la date t_i.
      Doit prendre comme argument la valeur S_{t_i} du sous-jacent.
    """
    q_up = (1 + r - d) / (u - d)
    q_down = 1 - q_up
    
    def fonction_prix_t_i(S):
        valeur = (fonction_prix_t_i_plus_one(S*u)*q_up + fonction_prix_t_i_plus_one(S*d)*q_down) / (1+r)
        return valeur
    
    if delta_flag:
            def delta_t_i(S):
                delta = (fonction_prix_t_i_plus_one(S*u) - fonction_prix_t_i_plus_one(S*d)) / (S*(u - d))
                return delta
   
    if delta_flag == 0:
        return fonction_prix_t_i
    
    else:
        return fonction_prix_t_i, delta_t_i
```

Nous pouvons déjà tester, pour le problème à $n=1$ période, le pricing d'un call $\Psi(S) = (S-K)^+$ (considéré dans le cours).

```python
S_0 = 4
u_u = 2
d = 1/2
r = 1/4
K = 5

## On verifie la condition d'absence d'arbitrage sur le marché contenant S et le taux r:
assert(d < 1 + r < u_u)
```

Calculons le prix en $t_0 = 0$ de l'option $\Psi(S_{t_1})$, et son delta:

```python
def payoff_call(S, strike = K):
    return np.maximum(S - strike, 0)

fonction_prix, fonction_delta = recursion(r, u_u, d, payoff_call, delta_flag=1)

prix = fonction_prix(S_0)
delta = fonction_delta(S_0)

print("Prix en t_0 du call de payoff (S_{t_1} - %1.1f)^+: %1.2f" %(K, prix))

print("Delta en t_0 du call de payoff (S_{t_1} - %1.1f)^+: %1.2f" %(K, delta))
```

### Cas à $n$ dates

Dans le cas $n$ dates $t_1, \dots, t_n$, on peut procéder de manière rétrograde pour calculer le prix et le delta en $t_0$ (et à toute date $t_i$, en fait) d'un payoff $\Psi(S_{t_n})$.

On pourra donc faire appel à la fonction prix_recursion (en appliquant le nombre approprié d'itérations).

```python
def recursion_n_i(r, u, d, payoff, i, n, delta_flag=0):
    """
    On effectue les itérations de la date t_n à à la date t_i, i < n,
    pour l'option de maturité t_n et de payoff = payoff(S_{t_n})
    
    Output: une fonction Python, la fonction de prix v(t_i, .) à la date t_i.
            Cette fonction prendra comme argument la valeur S_{t_i} du sous-jacent.
            
    Output:
    + Une fonction Python, la fonction de prix v(t_i, .) à la date t_i.
            
    + Si delta_flag: on renvoie la fonction de prix v(t_i, .) ET le delta(t_i, .)
      du portefeuille de couverture à la date t_i.
      Ces fonctions prendront comme argument la valeur S_{t_i} du sous-jacent.
    
    """
    fonction_prix_et_delta = [payoff, 0]
        
    for j in range(n, i, -1):
        fonction_prix_et_delta = recursion(r, u, d, fonction_prix_et_delta[0], delta_flag=1)
        ## la variable fonction_prix_et_delta contient un couple de fonctions
    
    if delta_flag == 0:
        return fonction_prix_et_delta[0]
    
    else:
        return fonction_prix_et_delta
```

Toujours pour le payoff  $\Psi(S) = (S-K)^+$ défini plus haut, on peut maintenant calculer le prix et le delta à n'importe quelle date $t_i$ pour une maturité $t_n > t_i$:

```python
n = 13
i = 0
S_ti = 99.12; u = 1.0514; d = (1.0-0.0489); r = 0.0031; K = 99.22

assert(d < 1 + r < u)

def payoff_call(S, strike = K):
    return np.maximum(S- strike, 0)

fonction_prix, fonction_delta = recursion_n_i(r, u, d, payoff_call, i, n, delta_flag=1)

prix = fonction_prix(S_ti)
delta = fonction_delta(S_ti)

print("Prix en t_%1.0f du call de payoff (S_{t_%1.0f} - %1.2f)^+ lorsque S_{t_%1.0f} = %1.3f: %1.3f" %(i, n, K, i, S_ti, prix))

print("Delta en t_%1.0f du call de payoff (S_{t_%1.0f} - %1.2f)^+ lorsque S_{t_%1.0f} = %1.3f: %1.3f" %(i, n, K, i, S_ti, delta))
```

```python
n = 8
i = 0

S_ti = 100.36; u = 1.1335; d = 1.0/1.1178; r = 0.0088; K = 100.40

assert(d < 1 + r < u)

def payoff_call(S, strike = K):
    return np.maximum(S - K, 0)

fonction_prix, fonction_delta = recursion_n_i(r, u, d, payoff_call, i, n, delta_flag=1)

prix = fonction_prix(S_ti)
delta = fonction_delta(S_ti)

print("Prix en t_%1.0f du put de payoff (S_{t_%1.0f} - %1.2f)^+ lorsque S_{t_%1.0f} = %1.3f: %1.3f" %(i, n, K, i, S_ti, prix))

print("Delta en t_%1.0f du put de payoff (S_{t_%1.0f} - %1.2f)^+ lorsque S_{t_%1.0f} = %1.3f: %1.3f" %(i, n, K, i, S_ti, delta))
```

### 1.1 Vérifier que le prix comptant du payoff $S_{t_n}$ dans ce modèle est bien égal à $S_0$:

```python
i = 0

def payoff_forward(S):
    return S

fonction_prix = recursion_n_i(r, u, d, payoff_forward, i, n)
prix = fonction_prix(S_0)

print("Prix en t_%1.0f de S_{t_%1.0f} lorsque S_0 vaut %1.2f: %1.2f" %(i, n, S_0, prix))
```

### 1.2 Afficher la fonction de prix $v(t_0, S_0)$ d'un call $(S_{t_n} - K)^+$ de strike et maturité donnés en fonction de la valeur de $S_0$

Nous pouvons également afficher le delta de l'option: nous allons calculer prix et delta sur la meme grille de valeur du spot $S_0$.

```python
## Une grille de 50 valeurs du sous-jacent S_0 entre K/3 et 2*K
nb_points = 50
spots = np.linspace(K/3, 2*K, nb_points)

call_prices = np.zeros(nb_points)
deltas = np.zeros(nb_points)

n = 10
i = 0

## Le payoff du call de strike K = 120 a été défini plus haut 
fonction_prix, fonction_delta = recursion_n_i(r, u, d, payoff_call, i, n, delta_flag=1)

for j, S in enumerate(spots):   
    call_prices[j] = fonction_prix(S)
    deltas[j] = fonction_delta(S)
```

Nous pouvons maintenant afficher les prix calculés en fonction de la valeur du spot $S_0$, ainsi que la borne inférieure pour un prix de call (valide dans n'importe quel modèle):
$$
C( (S_{t_n} - K)^+, t_n) \ge C( S_{t_n} - K, t_n) = S_{t_0} - K B(t_0, t_n)
= S_{t_0} - \frac{K}{(1+r)^{t_n}}
$$

```python
import matplotlib.pyplot as plt

plt.figure()
## On affiche les prix en fonction des valeurs du spot
plt.plot(spots, call_prices, color="b", label="Call price")

## On affiche la borne inf en fonction du spot
borne_inf = np.maximum(spots - K/(1+r)**n, 0) 

plt.plot(spots, borne_inf, color="k", label=u"Borne inf $(S - K/(1+r)^n)^+$")

plt.xlabel("spot S", fontsize=12)
plt.ylabel("price", fontsize=12)

plt.legend(loc="best", fontsize=12)

###################################################################
## Dans une autre figure, on affiche le delta en fonction du spot
###################################################################
plt.figure()

plt.plot(spots, deltas, color="g", label="Delta du call")
plt.axhline(1.0, color="k")

plt.xlabel("spot S", fontsize=12)
plt.ylabel("delta", fontsize=12)

plt.legend(loc="best", fontsize=12)
plt.show()
```

### 1.3 Alternativement, calculer les prix du call $(S_{t_n} - K)^+$ pour différents prix d'exercice $K$ (et une meme maturité $t_n$) et afficher les prix en fonction de leur strike.

```python
## Une grille de 50 strikes entre S_0/3 et 3*S_0
nb_points = 50
strikes = np.linspace(S_0/3, 3*S_0, nb_points)

call_prices = np.zeros(nb_points)

## On calcule les prix en t_0 = 0 des calls correspondants
n = 10
i = 0

for j, K in enumerate(strikes):
    
    def payoff_call(S):
        return np.maximum(S - K, 0)

    fonction_prix = recursion_n_i(r, u, d, payoff_call, i, n)

    call_prices[j] = fonction_prix(S_0)
```

```python
import matplotlib.pyplot as plt

## On affiche les prix en fonction des strikes
plt.plot(strikes, call_prices, color="b", label="Call price")

## On affiche la borne inf en fonction des strikes
borne_inf = np.maximum(S_0 - strikes/(1+r)**n, 0) 

plt.plot(strikes, borne_inf, color="k", label=u"Borne inf $(S_0 - K/(1+r)^n)^+$")

plt.xlabel("strike K", fontsize=12)
plt.ylabel("price", fontsize=12)

plt.legend(loc="best", fontsize=12)
```
